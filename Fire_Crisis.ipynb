```python
import pandas as pd

# Import the test and train datasets into pandas dataframe
df_train = pd.read_csv('train.csv', header=0, sep=',', quotechar='"')
df_test = pd.read_csv('test.csv', header=0, sep=',', quotechar='"')

df_train.head()
```

```python
for column in list(df_train):
  if df_train[str(column)].isnull().any():
    df_train[str(column)] = df_train[str(column)].fillna(df_train[str(column)].median())
  else:
    pass
```

```python
for column in list(df_train):
  print(df_train[str(column)].isnull().any())
```

```python
import numpy as np
msk = df_train.dtypes == object

for i in range(len(msk)):
  if msk[i] == True:
    list_object.append()
```

```python
list_object = []
list_number = []
df_number = pd.DataFrame()
df_object = pd.DataFrame()
for item in list(df_train):
  if df_train[item].dtype == object:
    list_object.append(item)
  else:
    list_number.append(item)
for col_num in list_number:
  df_number = pd.concat([df_number,df_train[col_num]],axis=1)
for col_obj in list_object:
  df_object = pd.concat([df_object,df_train[col_obj]],axis=1)

df_number.head()
df_object.head()
```

```python
df_var1 = pd.get_dummies(df_object.var1, prefix='var1')
df_var2 = pd.get_dummies(df_object.var2, prefix='var2')
df_var3 = pd.get_dummies(df_object.var3, prefix='var3')
df_var4 = pd.get_dummies(df_object.var4, prefix='var4')
df_var5 = pd.get_dummies(df_object.var5, prefix='var5')
df_var6 = pd.get_dummies(df_object.var6, prefix='var6')
df_var7 = pd.get_dummies(df_object.var7, prefix='var7')
df_var8 = pd.get_dummies(df_object.var8, prefix='var8')
df_dummy = pd.get_dummies(df_object.dummy, prefix='dummy')
df_number = pd.concat([df_number,df_var1,df_var2,df_var3,df_var4,df_var5,df_var6,df_var7,df_var8,df_dummy],axis=1)
df_number.head()
```
